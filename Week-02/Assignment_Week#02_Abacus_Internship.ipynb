```python
# Part 3: Machine Learning
# Task 1: Data Preparation
from sklearn.model_selection import train_test_split
InputFeature=dataset.drop("Species", axis=1)
OutputLabel=dataset["Species"]

InputFeatureTrain, InputFeatureTest, OutputLabelTrain, OutputLabelTest=train_test_split(InputFeature, OutputLabel, test_size=0.2, random_state=42)
print("Size of training data: ", len(InputFeatureTrain))
print("Size of testing data: ", len(InputFeatureTest))
```

```python
featuresOfDataset=["The_Length_of_Sepal", "The_Width_of_Sepal", "The_Length_of_Petal", "The_Width_of_Petal"]
print("Minimum Values:\n", dataset[featuresOfDataset].min())
print("\nMaximum Values:\n", dataset[featuresOfDataset].max())
print("\nAverage Values:\n", dataset[featuresOfDataset].mean())
```

```python
labels=dataset["Species"].unique()
print("Labels in Dataset are:\n", labels)
```

```python
from sklearn.preprocessing import LabelEncoder
encoderToEncodeTheValuesOfSpecies=LabelEncoder()
dataset["Species"]=encoderToEncodeTheValuesOfSpecies.fit_transform(dataset["Species"])
dataset
```

```python
# Task 2: Regression Model 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
InputFeaturesForLinearRegression=dataset[[
    "The_Length_of_Sepal",
    "The_Width_of_Sepal",
    "The_Length_of_Petal"]]
OutputFeatureForLinearRegression=dataset[[
    "The_Width_of_Petal"]]
InputFeaturesForLinearRegression_train, InputFeaturesForLinearRegression_test, OutputFeatureForLinearRegression_train, OutputFeatureForLinearRegression_test=train_test_split(InputFeaturesForLinearRegression, OutputFeatureForLinearRegression, 
                                                       test_size=0.2, random_state=42)
model=LinearRegression()
model.fit(InputFeaturesForLinearRegression_train, OutputFeatureForLinearRegression_train)
OutputFeaturesPredict=model.predict(InputFeaturesForLinearRegression_test)
print(OutputFeaturesPredict)
```

```python
comparison=InputFeaturesForLinearRegression_test.reset_index(drop=True)
comparison["Actual Value"]=OutputFeatureForLinearRegression_test.reset_index(drop=True)
comparison["Predicted Value"]=OutputFeaturesPredict
print("Actual vs Predicted Values:")
print(comparison)
```

```python
import numpy as np
mae=mean_absolute_error(OutputFeatureForLinearRegression_test, OutputFeaturesPredict)
print("Meam Asolute Error:     ", mae)
mse=mean_squared_error(OutputFeatureForLinearRegression_test, OutputFeaturesPredict)
print("Mean Squared Error:     ", mse)
rmse=np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
r2=r2_score(OutputFeatureForLinearRegression_test, OutputFeaturesPredict)
print("R² Score:               ", r2)
```

```python
# Task 3: Classification Model 
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
InputFeaturesForLogisticRegression=dataset[[
    "The_Length_of_Sepal",
    "The_Width_of_Sepal",
    "The_Length_of_Petal",
    "The_Width_of_Petal"]]
OutputFeatureForLogisticRegression=dataset[[
    "Species"]]

InputFeaturesForLogisticRegression_train, InputFeaturesForLogisticRegression_test, OutputFeatureForLogisticRegression_train, OutputFeatureForLogisticRegression_test=train_test_split(
InputFeaturesForLogisticRegression, OutputFeatureForLogisticRegression,test_size=0.2, random_state=42
)
model=LogisticRegression()
model.fit(InputFeaturesForLogisticRegression_train, OutputFeatureForLogisticRegression_train)
PredictedForLogistic=model.predict(InputFeaturesForLogisticRegression_test)
print(PredictedForLogistic)
```

```python
comparisonLogR=InputFeaturesForLogisticRegression_test.reset_index(drop=True)
comparisonLogR["Actual Species"]=OutputFeatureForLogisticRegression_test.reset_index(drop=True)
comparisonLogR["Predicted Species"]=PredictedForLogistic
print(comparisonLogR)
```

```python
accuracy = accuracy_score(OutputFeatureForLogisticRegression_test,PredictedForLogistic)
print("Accuracy: ", accuracy)

precision = precision_score(
    OutputFeatureForLogisticRegression_test,
    PredictedForLogistic,
    average="weighted"
)
print("Precision:", precision)

recall = recall_score(
    OutputFeatureForLogisticRegression_test,
    PredictedForLogistic,
    average="weighted"
)
print("Recall:   ", recall)

f1 = f1_score(
    OutputFeatureForLogisticRegression_test,
    PredictedForLogistic,
    average="weighted"
)
print("F1 Score: ", f1)

print("\nConfusion Matrix:")
print(confusion_matrix(OutputFeatureForLogisticRegression_test,PredictedForLogistic))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
InputFeaturesForDT=dataset[[
    "The_Length_of_Sepal",
    "The_Width_of_Sepal",
    "The_Length_of_Petal",
    "The_Width_of_Petal"]]
OutputFeatureForDT=dataset[[
    "Species"]]
InputFeaturesForDT_train, InputFeaturesForDT_test, OutputFeatureForDT_train, OutputFeatureForDT_test=train_test_split(
InputFeaturesForDT, OutputFeatureForDT,test_size=0.2, random_state=42
)
model=DecisionTreeClassifier()
model.fit(InputFeaturesForDT_train, OutputFeatureForDT_train)
predictedForDT=model.predict(InputFeaturesForDT_test)
print(predictedForDT)
```

```python
comparisonDT=InputFeaturesForDT_test.reset_index(drop=True)
comparisonDT["Actual Species"]=OutputFeatureForDT_test.reset_index(drop=True)
comparisonDT["Predicted Species"]=predictedForDT
print(comparisonDT)
```

```python
accuracy = accuracy_score(OutputFeatureForDT_test,predictedForDT)
print("Accuracy: ", accuracy)

precision = precision_score(
    OutputFeatureForDT_test,
    predictedForDT,
    average="weighted"
)
print("Precision:", precision)

recall = recall_score(
    OutputFeatureForDT_test,
    predictedForDT,
    average="weighted"
)
print("Recall:   ", recall)

f1 = f1_score(
    OutputFeatureForDT_test,
    predictedForDT,
    average="weighted"
)
print("F1 Score: ", f1)

print("\nConfusion Matrix:")
print(confusion_matrix(OutputFeatureForDT_test,predictedForDT))
```

```python
from sklearn.cluster import KMeans
inputFeaturesForKmeans=dataset[
    [
        "The_Length_of_Sepal",
        "The_Width_of_Sepal",
        "The_Length_of_Petal",
        "The_Width_of_Petal"
    ]
]
kmeans=KMeans(n_clusters=3, random_state=42)
kmeans.fit(inputFeaturesForKmeans)
dataset["Cluster"]=kmeans.labels_
print(dataset.head())
plt.scatter(
    dataset["The_Length_of_Petal"],
    dataset["The_Width_of_Petal"],
    c=dataset["Cluster"]
)
plt.title("K-Means Clustering")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
