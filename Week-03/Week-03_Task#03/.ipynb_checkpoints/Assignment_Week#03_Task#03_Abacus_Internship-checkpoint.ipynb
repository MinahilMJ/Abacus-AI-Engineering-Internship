```python
import os
import chromadb
from sentence_transformers import SentenceTransformer
```

```python
documents_folder="documents"
documents={}

for filename in os.listdir(documents_folder):
    if filename.endswith(".txt"):
        file_path=os.path.join(documents_folder,filename)
        with open(file_path,"r",encoding="utf-8") as file:
            documents[filename]=file.read()

print("Documents Loaded Successfully!\n")

for name,content in documents.items():
    print(f"File: {name}")
    print(f"Characters: {len(content)}")
    print("-"*50)
```

```python
chunks=[]

for filename,content in documents.items():
    paragraphs=content.split("\n\n")
    for paragraph in paragraphs:
        paragraph=paragraph.strip()
        if paragraph:
            chunks.append({"file":filename,"text":paragraph})

print(f"Total Chunks: {len(chunks)}")
```

```python
model=SentenceTransformer("all-MiniLM-L6-v2")

texts=[chunk["text"] for chunk in chunks]

embeddings=model.encode(texts)

print(f"Generated {len(embeddings)} embeddings.")
print(f"Embedding Dimension: {len(embeddings[0])}")
```

```python
client=chromadb.Client()

collection=client.get_or_create_collection(name="documents")

ids=[str(i) for i in range(len(chunks))]
documents_list=[chunk["text"] for chunk in chunks]
metadatas=[{"source":chunk["file"]} for chunk in chunks]

collection.add(
    ids=ids,
    documents=documents_list,
    embeddings=embeddings.tolist(),
    metadatas=metadatas
)

print("Vectors stored successfully!")
print(f"Total Documents: {collection.count()}")
```

```python
query=input("Enter your query: ")

query_embedding=model.encode([query]).tolist()

results=collection.query(
    query_embeddings=query_embedding,
    n_results=3
)

print("\nTop 3 Similar Chunks:\n")

for i in range(len(results["documents"][0])):
    print(f"Result {i+1}")
    print(f"Source: {results['metadatas'][0][i]['source']}")
    print(results["documents"][0][i])
    print("-"*50)
```

```python
query=input("Enter your query: ")

query_embedding=model.encode([query]).tolist()

results=collection.query(
    query_embeddings=query_embedding,
    n_results=3
)

print("\nTop Results:\n")

for i in range(len(results["documents"][0])):
    print(f"Result {i+1}")
    print(f"Source: {results['metadatas'][0][i]['source']}")
    print(f"Distance: {results['distances'][0][i]}")
    print(results["documents"][0][i])
    print("-"*50)
```
