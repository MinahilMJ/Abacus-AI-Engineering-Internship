```python
# 4. Functions
# Part A: Custom Map
numbers=[1,2,3,4,5]
def customMap(function, numbers):
    output=[]
    for num in numbers:
        output.append(function(num))
    return output
def squares(num):
    return num*num
output=customMap(squares,numbers)
print("Square of each number is the list is: ", output)
```

```python
# 4. Functions
# Part B: Custom Filter
# filter even no.s
numbers=[1,2,3,4,5,6]
def customFilter(function, numbers):
    output=[]
    for num in numbers:
        if function(num):
            output.append(num)
    return output
def even(num):
    return num%2==0
output=customFilter(even,numbers)
print("Even Numbers from the list are: ", output)
```

```python
# 4. Functions
# Part C: Custom Reduce
# add is used to reduce this given list
numbers=[1,2,3,4,5,6]
def customReduce(function, numbers):
    output=numbers[0]
    for num in numbers[1:]:
        output=function(output,num)
    return output
def add(x,y):
    return x+y
output=customReduce(add,numbers)
print("Sum of No's in the list is: ", output)
```

```python
# 5. File Handling
import os
import json

def WordsInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            content=file.read()
            words=content.split()
            return len(words)
    except Exception:
        return 0

def LinesInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            lines=file.readlines()
            return len(lines)
    except Exception:
        return 0

def FrequentWordsInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            freqWords=file.read()
            freqWords=freqWords.lower()
            freqWords=freqWords.split()
            freq={}
            for word in freqWords:
                word = word.strip('[]{}().,?!";')
                if word:
                    freq[word]=freq.get(word,0)+1
            sortedWords = sorted(freq.items(), key=lambda x: x[1], reverse=True)
            top10 = sortedWords[:10]
            return top10 if top10 else "None"
    except Exception:
        return "None"
                     
def main():
    path=r"C:\Users\Lenovo\Documents\Abacus Internship\Week-01\MiniAutomationTask\data"
    count=0
    file_data={}
    listOfFiles=[]
    total_words=0
    total_lines=0
    for file_name in os.listdir(path):
        if file_name.endswith(".txt"):
            #print("File: ",file_name)
            listOfFiles.append(file_name)
            full_path = os.path.join(path, file_name)
            
            #functionCalls
            count+=1
            word_count=WordsInEachFile(full_path) 
            total_words+=word_count

            line_count=LinesInEachFile(full_path)
            total_lines+=line_count

            freq_wordss=FrequentWordsInEachFile(full_path)
            
            file_data[file_name]={
                "word_count": word_count,
                "line_count": line_count,
                "frequent_words": freq_wordss
            }
            print("Total files:", count)
            print("Total words:", total_words)
            print("Total lines:", total_lines)
            print()
            for file, data in file_data.items():
                print(f"File: {file}")
                print(f"  Word Count: {data['word_count']}")
                print(f"  Line Count: {data['line_count']}")
                print(f"  Frequent Words: {data['frequent_words']}")
                print()
            with open("summaryOfFileHandling.json", "w", encoding="utf-8") as f:
                json.dump(file_data, f, indent=4)
if __name__=='__main__':
    main()
```

```python
# 6. Exception Handling
#storing errors in a json file
import json
from datetime import datetime

def logsOfErrors(expression, errorMessage):
    data={
        "expression": expression,
        "error": errorMessage,
        "time": str(datetime.now())
    }
    try:
        with open("errorLogsForCalculator.json", "r") as f:
            logs=json.load(f)
    except:
        logs=[]
    logs.append(data)
    with open("errorLogsForCalculator.json", "w") as f:
        json.dump(logs, f, indent=4)
def calculate(expression:str):
    allowedChars="0123456789+-%/*(). "
    for char in expression:
        if char not in allowedChars:
            return "Invalid Input. Alphabets not allowed. Enter Numbers."
    storeItems=[]
    i=0
    while i<len(expression):
        if expression[i]==" ":
            i+=1
        elif expression[i] in "0123456789.":
            specificNo=""
            while i < len(expression) and expression[i] in "0123456789.":
                specificNo+=expression[i]
                i+=1
            try:
                if "." in specificNo:
                    number=float(specificNo)
                else:
                    number=int(specificNo)
                storeItems.append(number)
            except ValueError:
                return "Invalid Expression Format"
        elif expression[i:i+2]=="**":
            storeItems.append("**")
            i+=2
        elif expression[i]=="-" and (i==0 or expression[i-1] in "+-*/%("):
            #for negative values first
            i+=1
            specificNo="-"
            while i<len(expression) and expression[i] in "0123456789.":
                specificNo+=expression[i]
                i+=1
            try:
                if "." in specificNo:
                    number=float(specificNo)
                else:
                    number=int(specificNo)
                storeItems.append(number)
            except ValueError:
                return "Invalid Expression Format"
        else:
            storeItems.append(expression[i])
            i+=1

    #BODMAS for ()
    while "(" in storeItems:
        begin=0
        for j in range (len(storeItems)):
            if storeItems[j]=="(":
                begin=j
        end=begin
        while end<len(storeItems) and storeItems[end]!=")":
            end+=1
        if end>=len(storeItems):
            return "Invalid Expression"
        subExpression=storeItems[begin+1:end]
        if not subExpression:
            return "Empty Brackets without values cannot be accepted"
        for groupOfOper in [["**"], ["/", "*", "%"], ["+", "-"]]:
            j=0
            while j<len(subExpression):
                if subExpression[j] in groupOfOper:
                    if j-1<0 or j+1>=len(subExpression):
                        return "Invalid Expression"
                    left=subExpression[j-1]
                    operator=subExpression[j]
                    right=subExpression[j+1]
                    if operator=="/" and right==0:
                        return "Division by Zero not possible"
                    if operator=="%" and right==0:
                        return "Modulo by Zero not possible"
                    if operator=="**":
                        result=left**right
                    elif operator=="/":
                        result=left/right
                    elif operator=="*":
                        result=left*right
                    elif operator=="%":
                        result=left%right
                    elif operator=="+":
                            result=left+right
                    elif operator=="-":
                            result=left-right
                    subExpression[j-1:j+2]=[result]
                else:
                    j+=1
        storeItems[begin:end+1]=subExpression
            
    #BODMAS for without ()    
    for groupOfOper in [["**"], ["/", "*", "%"], ["+", "-"]]:
        j=0
        while j<len(storeItems):
            if storeItems[j] in groupOfOper:
                if j-1<0 or j+1>=len(storeItems):
                    return "Invalid Expression"
                left=storeItems[j-1]
                operator=storeItems[j]
                right=storeItems[j+1]
                if operator=="/" and right==0:
                    return "Division by Zero not possible"
                if operator=="%" and right==0:
                    return "Modulo by Zero not possible"
                if operator=="**":
                    result=left**right
                elif operator=="/":
                    result=left/right
                elif operator=="*":
                    result=left*right
                elif operator=="%":
                    result=left%right
                elif operator=="+":
                        result=left+right
                elif operator=="-":
                        result=left-right
                storeItems[j-1:j+2]=[result]
            else:
                j+=1
    if len(storeItems)!=1:
        return "Invalid Expression"
    return storeItems[0]
            
if __name__=='__main__':
    print("Simple Calculator that operates on the principle of BODMAS")
    while True:
        try:
            ExpressionInput=input("Enter the Expression or type exit to leave")
            if ExpressionInput.strip().lower()=="exit":
                break
            print(f"Output: {calculate(ExpressionInput)}\n")
        except Exception as e:
            logsOfErrors(ExpressionInput, f"Runtime Error: {str(e)}")
            print("Output: Error occurred while processing the expression\n")
```

```python
# 7. OOP Challenges 
class Employee:
    def __init__(self, name, age, salary):
        self.name=name
        self.age=age
        self.salary=salary
    def job(self):
        return "They are doing their assigned work"
    def __str__(self):
        return f"Name of the Employee is {self.name}, their age is {self.age} and their salary is {self.salary}"

class Manager(Employee):
    def __init__(self, name, age, salary, department):
        super().__init__(name, age, salary)
        self.department=department
    def job(self):
        return "They are responsible for managing the employees"
    def __str__(self):
        return f"{super().__str__()}, in the department of {self.department}"

class Developer(Employee):
    def __init__(self, name, age, salary, language):
        super().__init__(name, age, salary)
        self.language=language
    def job(self):
        return "They are responsible for the writing code"
    def __str__(self):
        return f"{super().__str__()}, and they write code in {self.language}"

if __name__=="__main__":
    e=Employee("Maya", 30, 70000)
    m=Manager("Rehan", 50, 1000000, "IT")
    d=Developer("Minahil", 23, 120000, "Python")
    print(e)
    print(e.job())
    print(m)
    print(m.job())
    print(d)
    print(d.job())
```

```python
# 8. Python Problems 
# Part A: Valid Parentheses 
def checkPar(string1):
    stack=[]
    pairsOfParentheses={')':'(',']':'[','}':'{'}
    for i in string1:
        if i in "([{":
            stack.append(i)
        else:
            if not stack or stack[-1]!=pairsOfParentheses[i]:
                return False
            stack.pop()
    return len(stack)==0

if __name__=="__main__":
    string1=input("Enter Parentheses to check if valid or not")
    check=checkPar(string1)
    print("Check Proves: ",check)
```

```python
# 8. Python Problems 
# Part B: Longest Common Prefix 
def longestCommonPrefix(string1):
    if not string1:
        return ""
    prefix=string1[0]
    for word in string1[1:]:
        while not word.startswith(prefix):
            prefix=prefix[:-1]
    return prefix

if __name__=="__main__":
    string1=input("Enter words that are seperated by space")
    string1=string1.split()
    output=longestCommonPrefix(string1)
    print("Longest Common Prefix is: ", output)
```

```python
#10. Mini Automation Task
import os
import json

dictionaryOfTopWords={
    "artificial":True, 
    "intelligence":True, 
    "ai":True, 
    "deep":True, 
    "learning":True, 
    "machine":True, 
    "algorithm":True, 
    "linear":True, 
    "logistic":True, 
    "python":True
}

def WordsInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            content=file.read()
            words=content.split()
            return len(words)
    except Exception:
        return 0

def CharacterCountInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            characters=file.read()
            return len(characters)
    except Exception:
        return 0

def LinesInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            lines=file.readlines()
            return len(lines)
    except Exception:
        return 0

def TopWordsInEachFile(full_path):
    try:
        with open(full_path, "r", encoding="utf-8") as file:
            topWords=file.read()
            topWords=topWords.lower()
            topWords=topWords.split()

            Top=[]
            
            for word in topWords:
                Words = word.strip('[]{}().,?!";')
                if Words in dictionaryOfTopWords and Words not in Top:
                    Top.append(Words)
            if not Top:
                return "None"

            return Top
    except Exception:
        return "None"
                     
def main():
    path=r"C:\Users\Lenovo\Documents\Abacus Internship\Week-01\MiniAutomationTask\data"
    count=0
    file_data={}
    listOfFiles=[]
    total_words=0
    total_chars=0
    total_lines=0
    for file_name in os.listdir(path):
        if file_name.endswith(".txt"):
            #print("File: ",file_name)
            listOfFiles.append(file_name)
            full_path = os.path.join(path, file_name)
            
            #functionCalls
            count+=1
            word_count=WordsInEachFile(full_path) 
            total_words+=word_count

            char_count=CharacterCountInEachFile(full_path)
            total_chars+=char_count

            line_count=LinesInEachFile(full_path)
            total_lines+=line_count

            top_wordss=TopWordsInEachFile(full_path)
            
            file_data[file_name]={
                "word_count": word_count,
                "char_count": char_count,
                "line_count": line_count,
                "top_words": top_wordss
            }
    folderInfo={
        "total_documents": count,
        "total_word_count": total_words,
        "documents": file_data,          
    }
    print(json.dumps(folderInfo, indent=2, ensure_ascii=False))
if __name__=='__main__':
    main()
```
